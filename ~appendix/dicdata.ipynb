# 한국투자(kor_true)

```python
df = pd.read_csv('/home/eunhyea/EARTH/ConceptMap/dicdata/한국투자.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df.rename(columns={'word':'용어', 'content':'설명'}, inplace=True)

df.dropna(inplace=True)
```

```python
import pandas as pd
import re
import ast

meta_keys = ['한자', '영문', '동의어', '참조어', '외래어']

def split_meta_and_description(text):
    if pd.isna(text):
        return [], ''

    lines = text.splitlines()
    details = []
    description_lines = []

    for line in lines:
        line = line.strip()
        matched = False
        for key in meta_keys:
            if line.startswith(f"{key} :"):
                value = line.split(":", 1)[1].strip()
                value = [v.strip() for v in value.split(',')]
                details.append({key:value})
                matched = True
                break
        if not matched:
            description_lines.append(line)

    cleaned_description = ' '.join(description_lines)
    return details, cleaned_description


# 적용
details_list = []
cleaned_descriptions = []

for text in df['설명']:
    details, cleaned = split_meta_and_description(text)
    details_list.append(details)
    cleaned_descriptions.append(cleaned)

df['details'] = details_list
df['설명'] = cleaned_descriptions

# 확인
df.head()
```

```python
def extract_kwdetails(row):

    # relation 값
    relations = []
    if row['details']:
        try:
            parsed = row['details'] # list 자료형
            if isinstance(parsed, list) and parsed:
                for el in parsed:
                    if not isinstance(el, dict):
                        continue
                    for key, value in el.items():
                        if key != '참조어':  # '참조어' 제외
                            relations.extend(value)
        except Exception as e:
            print(f"변환 오류: {e}, 값: {row['details']}")
        
        return set(relations)
    else : return None


# 새로운 컬럼 생성
df['KWdetails'] = df.apply(extract_kwdetails, axis=1)

# 확인
df.iloc[17][['KWdetails','details']]

df.drop(columns='details', inplace=True)

df.rename(columns={'KWdetails':'details'}, inplace=True)
```

```python
df.to_csv('한국투자_pre_details.csv', index=False)
```

# 국립국어원

```python
import numpy as np
import pandas as pd

import ast


df = pd.read_csv('/home/eunhyea/EARTH/ConceptMap/dicdata/국립국어원.csv', header=0)

df.replace({np.nan: ''}, inplace=True)

def extract_kwdetails(row):
    # origin 값
    origin = row['origin'] if pd.notna(row['origin']) else ''
    
    # translations 값에서 'translation'만 추출
    translations = ''
    if row['translations']:
        try:
            parsed = ast.literal_eval(row['translations'])
            if isinstance(parsed, list) and parsed:
                translations = parsed[0].get('translation', '').strip()
        except Exception as e:
            print(f"변환 오류: {e}, 값: {row['translations']}")

    # relation 값
    relations = []
    if row['related']:
        try:
            parsed = ast.literal_eval(row['related'])
            if isinstance(parsed, list) and parsed:
                for el in parsed:
                    if el.get('category', '') != '참고어':
                        val = el.get('values', '')
                        # 'values'가 set일 경우, 리스트로 풀어주기
                        if isinstance(val, set):
                            relations.extend(list(val))  # set -> list
                        elif isinstance(val, list):
                            relations.extend(val)       # 이미 리스트면 그대로
                        elif isinstance(val, str):
                            relations.append(val)       # 단일 문자열
        except Exception as e:
            print(f"변환 오류: {e}, 값: {row['related']}")

    result = []
    if origin:
        result.append(origin.strip())
    if translations:
        result.append(translations)
    if relations:
        result.extend(relations)
    
    # return ', '.join(set(result))
    return set(result)


# 새로운 컬럼 생성
df['KWdetails'] = df.apply(extract_kwdetails, axis=1)
df_pre = df[['korean', 'definition', 'KWdetails']].rename(columns={'korean':'용어', 'definition':'설명', 'KWdetails':'details'})
df_pre.to_csv('국립국어원_pre_details.csv',index=False)

df_pre = df[['korean', 'definition']].rename(columns={'korean':'용어', 'definition':'설명', 'KWdetails':'details'})
df_pre.to_csv('국립국어원_pre.csv', index=False)
```

# 중복값 처리

```python
import re

df_sorted = pd.read_csv('통합_ver1.csv', header=0)
# df_sorted.fillna('',inplace=True) # 모든 컬럼 문자열로 전처리함....

# 1) details 컬럼(문자열로 된 "['A','B',...]" 형태)을 실제 리스트로
def to_set(x):
    if isinstance(x, str):
        try:
            lst = ast.literal_eval(x)
            return set(lst) if isinstance(lst, (set, tuple)) else set()
        except:
            # "A, B, C" 처럼 리스트 리터럴이 아닐 때
            return [p.strip() for p in x.split(',') if p.strip()]
    elif isinstance(x, (set, tuple)):
        return set(x)
    else:
        return set()

df_sorted['details_list'] = df_sorted['details'].apply(to_set)
df_sorted
```

```python
# 2) 괄호 안 텍스트를 set에 add
def add_parenthesis(details: set, term: str) -> set:
    # details: 기존 set
    # term: '용어' 문자열, 예: '인공지능(AI: Artificial Intelligence)'
    m = re.search(r'\((?=\D)([^)]*)\)', term)
    if m:
        extra = m.group(1).strip()
        if extra:
            # 방법 1: in-place로 추가
            details.add(extra)
    return details

df_sorted['details_list'] = df_sorted.apply(
    lambda row: add_parenthesis(row['details_list'], row['용어']),
    axis=1
)
df_sorted
```

```python
# 3) 용어에서 괄호 제거
df_sorted['term_clean'] = df_sorted['용어'].str.replace(r'\([^)]*\)', '', regex=True).str.strip()

# 4) 최종 details_pre 리스트로 합치기 (term_clean + details_list)
df_sorted['details_pre_list'] = df_sorted.apply(
    lambda r: set(r['details_list']) | {r['term_clean']},
    axis=1
)

# 5) 띄어쓰기 제거 동의어도 추가
df_sorted['details_pre_list'] = df_sorted.apply(
    lambda r: set(r['details_pre_list']) 
              | { re.sub(r'\s+', '', r['term_clean']) } 
              if r['term_clean'] else set(r['details_pre_list']),
    axis=1
)

df_sorted
```

```python
df_sorted['details_pre'] = df_sorted['details_pre_list'].apply(lambda lst: ', '.join(lst))
df_synoynms = df_sorted.drop(columns=['details','details_list','term_clean'])
df_synoynms.to_csv('통합_ver2.csv', index=False)
```

```python
import pandas as pd
import ast
from collections import defaultdict

# --- 1. DataFrame 불러오기 (예시) ------------------------------------------
df = pd.read_csv('통합_ver2.csv', header=0)

# --- 2. 'details_pre' 컬럼에서 실제 set 형태로 꺼내기 ---------------------------
def parse_details(x):
    """
    details_pre 컬럼이
      - 리스트/셋 객체이면 그대로 set으로 변환,
      - 문자열이면 ',' 를 구분자로 분리해 set 생성,
      - 그 외(None, NaN)이면 빈 set 반환
    """
    if isinstance(x, (list, set)):
        return set(x)
    if not isinstance(x, str):
        return set()
    # 쉼표로 분리
    parts = [p.strip() for p in x.split(',')]
    # 빈 문자열 제거
    return set([p for p in parts if p]).union({p.lower() for p in parts if p})

# 적용
df['details_set'] = df['details_pre'].apply(parse_details)

# --- 3. Union–Find(Disjoint Set) 으로 동의어 그룹 만들기 ---------------------
parent = {}

def find(u):
    parent.setdefault(u, u)
    if parent[u] != u:
        parent[u] = find(parent[u])
    return parent[u]

def union(a, b):
    ra, rb = find(a), find(b)
    if ra != rb:
        parent[rb] = ra

# (A) 모든 용어(term)를 parent에 등록
for term in df['용어']:
    parent.setdefault(term, term)

# (B) detail → 용어 리스트 매핑
from collections import defaultdict
detail_to_terms = defaultdict(list)
for term, details in zip(df['용어'], df['details_set']):
    for d in details:
        detail_to_terms[d].append(term)

# (C) 같은 detail을 공유하는 용어들끼리 union
for detail, terms in detail_to_terms.items():
    # detail로 연결된 term 그룹이 2개 이상일 때만
    if len(terms) > 1:
        first = terms[0]
        for other in terms[1:]:
            union(first, other)
# ------------------------------------------------------------------------------

# --- 4. 각 용어가 속한 대표(root)를 찾아 'component' 컬럼에 기록 -------------
df['component'] = df['용어'].apply(find)
# -----------------------------------------------------------------------------


# --- 5. component별로 용어 그룹 매핑하기 -------------------------
from collections import defaultdict

# component(root) → 전체 그룹(용어 + 모든 details) 매핑
comp_to_group = defaultdict(set)
for term, details, comp in zip(df['용어'], df['details_set'], df['component']):
    # 1) 원래 용어도 추가
    comp_to_group[comp].add(term)
    # 2) 그 용어에 딸린 모든 details_set 아이템도 추가
    comp_to_group[comp].update(details)

# 각 행에 해당 component 의 전체 그룹을 set 으로 할당
df['details_group'] = df['component'].map(comp_to_group)

# --- 5. 그룹별로 대표 행만 골라내기 -------------------------
keep_idxs = []
for comp, group_df in df.groupby('component'):
    # 1) component(대표어)와 용어(term)가 같은 행이 있으면 그 행의 idx 사용
    is_canon = group_df['용어'] == comp
    if is_canon.any():
        idx = group_df[is_canon].index[0]
    else:
        # 2) 아니면 기존처럼 설명 길이 최댓값 행
        idx = group_df['설명'].str.len().idxmax()
    keep_idxs.append(idx)

# 최종 결과 DataFrame
df_dedup = df.loc[keep_idxs].drop(columns=['details_set', 'component', 'details_pre']).reset_index(drop=True)
# -------------------------------------------------------------------------

# --- 6. 결과 확인 ---------------------------------------------------------
print(f"원본 {len(df)}개 → 중복 제거 후 {len(df_dedup)}개")
df_dedup.info()
```
